### Standardization

	Standardize the numerical columns using the formula: z=x-mu/sigma

```python
standardize_columns1=df.select_dtypes(include=['int'])
standardize_columns1
```

```python
def standardize_columns1(int):
    mean = int.mean()
    std_dev = int.std()
    standardized_columns1 = (int - mean) / std_dev
    return standardized_columns1
```

```python
for int in numerical_columns1:
    df[int] = standardize_columns1(df[int])

df[numerical_columns1].head()
```

```python
standardize_columns2=df.select_dtypes(include=['float'])
standardize_columns2
```

```python
def standardize_columns2(float):
    mean = float.mean()
    std_dev = float.std()
    standardized_columns2 = (float - mean) / std_dev
    return standardized_columns2
```

```python
for float in numerical_columns2:
    df[float]=standardize_columns2(df[float])

df[numerical_columns2].head()
```
